````python
# 6.Write a code that generates the squares of numbers from 1 to n using a generator.

def generate_squares(n):
    for i in range(1, n + 1):
        yield i * i  # This will generate the square of i

# Example usage:
n = 5
for square in generate_squares(n):
    print(square)
````

````python
# 7.Write a code that generates palindromic numbers up to n using a generator.

def is_palindrome(num):
    return str(num) == str(num)[::-1]

def generate_palindromes(n):
    for i in range(1, n + 1):
        if is_palindrome(i):
            yield i  # Yield the palindromic number

# Example usage:
n = 100
for palindrome in generate_palindromes(n):
    print(palindrome)
````

````python
# 8.Write a code that generates even numbers from 2 to n using a generator.

def generate_even_numbers(n):
    for i in range(2, n + 1, 2):  # Start from 2, increment by 2
        yield i  # Yield the even number

# Example usage:
n = 20
for even_number in generate_even_numbers(n):
    print(even_number)
````

````python
# 9.Write a code that generates powers of two up to n using a generator.

def generate_powers_of_two(n):
    power = 1  # Start with 2^0
    while power <= n:
        yield power  # Yield the current power of two
        power *= 2  # Multiply by 2 for the next power

# Example usage:
n = 100
for power in generate_powers_of_two(n):
    print(power)
````

````python
# 10.Write a code that generates prime numbers up to n using a generator.

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

def generate_primes(n):
    for i in range(2, n + 1):
        if is_prime(i):
            yield i  # Yield the prime number

# Example usage:
n = 50
for prime in generate_primes(n):
    print(prime)
````

## 27. Explain the use of access modifiers in Python classes

Access modifiers in Python classes control the visibility and accessibility of class members (attributes and methods). They help encapsulate data and prevent unintended modifications.
Python has three types of access modifiers:


1.   **Public**:
Accessible from anywhere, both inside and outside the class. By default, all members are public unless specified otherwise.
2.   Protected:
Accessible within the class and its subclasses. Denoted by a single underscore prefix _. While accessible from outside the class, it signals that it should not be accessed directly.
3. Private:
Accessible only within the class. Denoted by a double underscore prefix __. Python applies name mangling to make it harder (but not impossible) to access from outside.





## 28. What is inheritance in OOP?

Types of Inheritance:


1.   Single Inheritance: A child class inherits from only one parent class.

4.  Hierarchical Inheritance: Multiple child classes inherit from a single parent class.
5.  Hybrid Inheritance: A combination of multiple types of inheritance.

## 30. Explain method overriding in Python

Method overriding in Python occurs when a subclass provides a different implementation for a method that is already defined in its parent class. This enables a subclass to modify or extend the behavior of an inherited method without changing the parent class definition. When a method is called on an object of the subclass, the overridden method in the subclass is executed instead of the original method in the parent class.

Example:

```
class Animal:
    def speak(self):
        print("Generic animal sound")

class Dog(Animal):
    def speak(self):
        print("Woof!")

class Cat(Animal):
    def speak(self):
        print("Meow!")

animal = Animal()
dog = Dog()
cat = Cat()

animal.speak()  # Output: Generic animal sound
dog.speak()     # Output: Woof!
cat.speak()     # Output: Meow!
```
In the example above, the Dog and Cat classes override the speak method of the Animal class. When speak is called on instances of Dog and Cat, their specific implementations are executed, demonstrating method overriding.

## 36. What is abstraction in Python? How is it implemented?
Abstraction in Python (and Object-Oriented Programming in general) means hiding internal implementation details and showing only essential features of an object. It helps reduce complexity and makes the code easier to manage and maintain.

How it is implemented?

Python uses the abc module (Abstract Base Classes) to implement abstraction.

✅ Steps:


1.   Import ABC and abstractmethod from the abc module.
2.   Define an abstract class by inheriting from ABC.
3.   Use the @abstractmethod decorator for methods that must be implemented by subclasses.

Example:


```
from abc import ABC, abstractmethod

# Abstract class
class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass  # No implementation here

# Concrete class
class Dog(Animal):
    def make_sound(self):
        print("Woof!")

# Concrete class
class Cat(Animal):
    def make_sound(self):
        print("Meow!")

# Instantiate concrete classes
dog = Dog()
cat = Cat()

dog.make_sound()  # Output: Woof!
cat.make_sound()  # Output: Meow!

```

## 39. How can you achieve abstraction using interfaces in Python?

✅ Steps to Achieve Abstraction via Interfaces in Python:
1. Import ABC and abstractmethod from the abc module.
2. Create an abstract class that inherits from ABC.
3. Declare abstract methods using the @abstractmethod decorator — these act as interface methods.
4. Subclasses must implement all abstract methods, or they will also be treated as abstract.

Example:Interface-like Abstraction in Python


```
from abc import ABC, abstractmethod

# Interface-like abstract class
class Vehicle(ABC):
    
    @abstractmethod
    def start_engine(self):
        pass

    @abstractmethod
    def stop_engine(self):
        pass

# Concrete class implementing the interface
class Car(Vehicle):
    def start_engine(self):
        print("Car engine started.")

    def stop_engine(self):
        print("Car engine stopped.")

# Trying to instantiate Vehicle will raise an error
# vehicle = Vehicle()  # ❌ TypeError

# Valid instantiation of subclass
my_car = Car()
my_car.start_engine()  # Output: Car engine started.
my_car.stop_engine()   # Output: Car engine stopped.

```




## 40. Can you provide an example of how abstraction can be utilized to create a common interface for a group of related classes in Python?

🧠 **Scenario**: You have multiple types of Payment methods — like CreditCard, PayPal, and UPI — and you want all of them to follow a common interface.

✅ Step-by-Step Code Example:

```

        pass

        print(f"Paid ₹{amount} using PayPal.")

class UPI(PaymentMethod):
    def pay(self, amount):
        print(f"Paid ₹{amount} using UPI.")

# Function that uses the common interface
def make_payment(payment_method: PaymentMethod, amount):
    payment_method.pay(amount)

# Usage

make_payment(PayPal(), 750)
make_payment(UPI(), 500)

```


## 41. How does Python achieve polymorphism through method overriding?

Python achieves polymorphism through method overriding by allowing subclasses to provide a specific implementation of a method that is already defined in their parent class. This means the same method name can behave differently based on the object calling it.

✅ Example of Polymorphism via Method Overriding


```
class Animal:
    def make_sound(self):
        print("Some generic animal sound")

class Dog(Animal):
    def make_sound(self):
        print("Woof!")

class Cat(Animal):
    def make_sound(self):
        print("Meow!")

# Function that takes any Animal and calls make_sound
def speak(animal):
    animal.make_sound()

# Different behaviors for the same method name
speak(Animal())  # Output: Some generic animal sound
speak(Dog())     # Output: Woof!
speak(Cat())     # Output: Meow!

```
🧾 Explanation:
* make_sound() is defined in the parent class Animal.

* Both Dog and Cat override this method with their own version.
* The speak() function demonstrates polymorphism by calling the overridden method appropriate to the object passed in.

## 44. How does polymorphism improve code readability and reusability?
Polymorphism improves code readability and reusability by allowing the same code to be used with different types of objects without needing to know their specific implementations. This leads to more concise and easier-to-understand code, as developers can write generic code that works with a variety of classe

## 45. Describe how Python supports polymorphism with duck typing.

In Python, duck typing means that the type or class of an object is less important than the methods and properties it defines. As long as an object implements the required behavior, it can be used anywhere, regardless of its actual class.

✅ How Duck Typing Enables Polymorphism in Python
* Python does not require inheritance or a common parent class to treat objects polymorphically.

* What matters is whether the object has the necessary methods or attributes.

* This is implicit polymorphism: different types of objects can be used interchangeably if they support the same operations.

🔧 Example: **Duck Typing in Action**


```
class Duck:
    def quack(self):
        print("Quack, quack!")

class Person:
    def quack(self):
        print("I'm pretending to be a duck!")

# Function using duck typing
def make_it_quack(thing):
    thing.quack()

# Works with any object that has a quack method
make_it_quack(Duck())     # Output: Quack, quack!
make_it_quack(Person())   # Output: I'm pretending to be a duck!

```
🧠 Key Points:
* make_it_quack() doesn’t care what the actual type of thing is.

* As long as the object has a quack() method, the function works — this is polymorphism via duck typing.

* No need to explicitly inherit from a base class or implement an interface.


## 46. How do you achieve encapsulation in Python?

Encapsulation in Python is achieved through access modifiers, which control the visibility of class members (attributes and methods). Python uses naming conventions to indicate the intended access level, rather than explicit keywords like public, private, or protected found in other languages.

🔐 Example: Encapsulation in Action

```
class Employee:
    def __init__(self, name, salary):
        self.name = name          # public
        self.__salary = salary    # private

    def get_salary(self):
        return self.__salary      # controlled access

    def set_salary(self, new_salary):
        if new_salary > 0:
            self.__salary = new_salary
        else:
            print("Invalid salary!")

# Create an employee object
emp = Employee("Alice", 50000)

# Accessing public member
print(emp.name)  # Alice

# Trying to access private member (will raise AttributeError)
# print(emp.__salary)  # ❌ Error

# Use getter and setter to access private data
print(emp.get_salary())  # ✅ 50000
emp.set_salary(60000)
print(emp.get_salary())  # ✅ 60000

```



## 47. Can encapsulation be bypassed in Python? If so, how?

Encapsulation in Python is a convention, not a strict rule. While it suggests that attributes prefixed with a single underscore (_) or double underscore (__) should be treated as non-public, it doesn't prevent access from outside the class. Therefore, encapsulation can be bypassed using name mangling.

Name mangling in Python transforms attributes with double underscores to _ClassName__attribute making it more difficult to access directly, but not impossible.




The __add__() method in Python is a special method, also known as a "magic" or "dunder" method, that enables operator overloading for the + operator. ## 65. Explain the usage of the __iter__() and __next__() methods in Python. Provide an example using iterators.

In Python, the __iter__() and __next__() methods are fundamental for implementing iterators.

__iter__():
* This method is called when iteration is initialized on an object. It must return the iterator object itself. In most cases, it simply returns self. This method enables the object to be used in a for loop or with the iter() function.

__next__():
* This method is responsible for returning the next item in the sequence. When there are no more items to return, it should raise the StopIteration exception. This signals the end of the iteration.

Example:


```
class Counter:
    def __init__(self, start, end):
        self.current = start
        self.end = end

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.end:
            raise StopIteration
        else:
            value = self.current
            self.current += 1
            return value

# Using the Counter iterator
my_counter = Counter(1, 5)
for num in my_counter:
    print(num)
# Output: 1 2 3 4 5
```
In this example, the Counter class is an iterator. The __iter__() method returns the object itself, and the __next__() method returns the next number in the sequence until it reaches the end, at which point it raises StopIteration.

## 70. How does encapsulation relate to property decorators in Python? Provide an example showcasing encapsulation using property decorators.

Encapsulation, a fundamental principle of object-oriented programming, involves bundling data (attributes) and methods that operate on that data within a single unit (a class). It also restricts direct access to some of the object's components, protecting data integrity. Property decorators in Python facilitate encapsulation by allowing controlled access to class attributes through following

* Getter (@property): Provides read-only access to an attribute.

* Setter (@property.setter): Allows controlled modification of an attribute.

* Deleter (@property.deleter): Defines behavior when an attribute is deleted.

🧩 Example: Encapsulation Using Property Decorators


```

        self.owner = owner
        self._balance = balance  # private variable

    def balance(self):
        # Encapsulation: Read-only access to balance
        return self._balance

    @balance.setter
    def balance(self, amount):
        # Encapsulation: Prevent setting a negative balance
        if amount < 0:
            print("Balance cannot be negative!")
        else:
            self._balance = amount

    @balance.deleter
    def balance(self):
        # Encapsulation: Prevent deletion of balance directly
        print("Cannot delete the balance attribute.")
        # Optionally, we can raise an exception or provide other logic here.

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
        else:
            print("Deposit amount must be positive.")
```
